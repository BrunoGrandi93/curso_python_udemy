## 9.3 Tratando dados (Exemplo prático 1)

A partir dos dados de jogadores da NBA (https://www.kaggle.com/drgilermo/nba-players-stats), pede-se:

```python
import pandas as pd

file = 'datasets_1358_30676_Players.csv'

df = pd.read_csv(file)
```

- Qual a média de altura e de peso dos jogadores?

```python
a = df['height'].mean()
print("A média de altura dos jogadores é de:", a , "centímetros.")
b = df['weight'].mean()
print("A média de peso dos jogadores é de:", b , "quilos.")
```

- Crie uma nova coluna com a razão peso/altura.

```python
razao = df['weight']/df['height']

df["razao"] = razao
df
```

- Qual a maior altura, o maior peso, a menor altura e menor peso?

```python
print("A maior altura dentre todos os jogadores é de:",df['height'].max(),"centímetros")
print("A menor altura dentre todos os jogadores é de:",df['height'].min(),"centímetros")
print("O maior peso dentre todos os jogadores é de:",df['weight'].max(),"quilogramas")
print("O menor peso dentre todos os jogadores é de:",df['weight'].min(),"quilogramas")
```

- Qual é o estado onde nasce o maior número de jogadores?

```python
df['birth_state'].value_counts()
```

- Qual o jogador mais alto?

```python
df.sort_values(by='height').tail(20)
```
