## 9.13 Desafio - regressão linear: correlação com dados reais

Faça uma regressão linear com os dados da NBA e encontre a correlação com o peso dos atletas.
- Encontre a correlaçao de peso e altura

```python
file = 'datasets_1358_30676_Players.csv'

import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
NBA_df = pd.read_csv(file)
NBA_df = NBA_df.dropna()
NBA_df
```

**Plot**

```python
plt.figure()
plt.plot(NBA_df['height'],NBA_df['weight'],'ro')
plt.ylabel('Peso')
plt.xlabel('Altura')
plt.grid()

plt.show()
```

**Regressão Linear**

```python
Y = NBA_df['weight']
X = NBA_df['height']
```

```python
estatistica = sm.OLS(Y,X).fit()

print(estatistica.summary())
```

```python
estatistica.predict()
```

```python
plt.figure()
plt.plot(NBA_df['height'],NBA_df['weight'],'ro')
plt.plot(NBA_df['height'],estatistica.predict(),'k--')
plt.ylabel('Peso')
plt.xlabel('Altura')
plt.grid()

plt.show()
```

- Encontre a correlação entre idade e altura

```python
NBA_df = pd.read_csv(file)
NBA_df = NBA_df.dropna()
NBA_df['age'] = 2025 - NBA_df['born']
NBA_df
```

```python
plt.figure()
plt.plot(NBA_df['age'],NBA_df['weight'],'ro')

plt.ylabel('Peso')
plt.xlabel('Idade')
plt.grid()

plt.show()
```

**Regressão Linear**

```python
Y = NBA_df['weight']
X = NBA_df['age']
```

```python
estatistica = sm.OLS(Y,X).fit()

print(estatistica.summary())
```

```python
estatistica.predict()
```

**Plot**

```python
plt.figure()
plt.plot(NBA_df['age'],NBA_df['weight'],'ro')
plt.plot(NBA_df['age'],estatistica.predict(),'k--')
plt.ylabel('Peso')
plt.xlabel('Idade')
plt.grid()

plt.show()
```

- Encontre a correlação com multiplas variáveis (peso com altura e idade)

**Regressão Linear Múltipla**

```python
Y = NBA_df['weight']
X = NBA_df[['age','height']]
```

```python
estatistica = sm.OLS(Y,X).fit()

print(estatistica.summary())
```

```python
estatistica.predict()
```

**Plot**

```python
plt.figure()
plt.plot(NBA_df['height'],NBA_df['weight'],'ro')
plt.plot(NBA_df['height'],estatistica.predict(),'ko')
plt.ylabel('Peso')
plt.xlabel('Altura')
plt.grid()

plt.show()
```

**Regressão Linear de Multiplos Fatores e Constante**

```python
Y = NBA_df['weight']
X = NBA_df[['age','height']]
X = sm.add_constant(X)
```

```python
estatistica = sm.OLS(Y,X).fit()

print(estatistica.summary())
```

```python
estatistica.predict()
```

**Plot**

```python
plt.figure()
plt.plot(NBA_df['height'],NBA_df['weight'],'ro')
plt.plot(NBA_df['height'],estatistica.predict(),'ko')
plt.ylabel('Peso')
plt.xlabel('Altura')
plt.grid()

plt.show()
```
