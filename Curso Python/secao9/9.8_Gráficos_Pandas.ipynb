## 9.8 Visualização de gráficos com Pandas


Nesta seção veremos:

- df.plot(x='x',y='y')
- df.plot.bar(x='x',y='y')
- df.plot.pie(subplots=True ou y='y')
- df.plot.scatter(x='x',y='y')


Acesse: https://pandas.pydata.org/docs/user_guide/visualization.html#visualization-hist

```python
import pandas as pd

import numpy as np

arr = np.arange(10)
arr
```

```python
serie_1 = pd.Series(arr)
serie_1
```

```python
serie_1.plot()
```

```python
serie_1 = pd.Series(arr.cumsum())

serie_1.plot()
```

```python
df1 = pd.DataFrame({'a':np.arange(50)**2,'b':np.arange(50)*3,'c':np.arange(50)*5})

df1.plot(x='c',y=['a','b'])
```

```python
file = 'Acoes_ficticias.csv'

df = pd.read_csv(file,delimiter=';')

df['valor_total'] = df.preco_compra*df.Quantidade

soma_qtd_valor = pd.pivot_table(df,index='Papel',values=['Quantidade','valor_total'],aggfunc='sum')
soma_qtd_valor

soma_qtd_valor['p_medio'] = soma_qtd_valor.valor_total/soma_qtd_valor.Quantidade

soma_qtd_valor.plot.bar()
```

```python
soma_qtd_valor.plot.bar(y='Quantidade')
```

```python
soma_qtd_valor.plot.pie(y='Quantidade')
```
