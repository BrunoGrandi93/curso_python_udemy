## 5.4 Exercício 1 - 
Crie uma função que receba uma frase e retorne quantas palavras tem a frase. Dica: Utilize o método split().

```python
def cont_palavras(frase):
    lista_palavras = frase.split(" ")
    numero = len(lista_palavras)
    return numero

a = cont_palavras("Minha mãe mandou eu escolher este daqui")
print(a)

frase = "Testando outra maneira de input de variáveis em funções"

b = cont_palavras(frase)
print(b)
```

## 5.5 Exercício 2 -
Crie uma função que lê um arquivo de extensão .txt e retorne quantas palavras tem esse arquivo. Dica: utilize o método readlines()

```python
def numero_palavras(arquivo_path):
    with open(arquivo_path) as f:
        string = f.read()
        f.close

    lista_palavras = string.replace('\n','')
    lista_palavras = lista_palavras.split(' ')
    numero_palavras = len(lista_palavras)

    return numero_palavras
    
nome_arquivo = 'Arquivo_sujo.txt'
numero_palavras(nome_arquivo)
```
